```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  mean_squared_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('newdata\\new_heart_data.csv')
```

```python
x = data['MaxHR'].values.reshape(-1, 1)
y = data['HeartDisease']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y, color ='green')
plt.plot(x, model.predict(x), color='red')
plt.xlabel('hr')
plt.ylabel('heart')
plt.title('fitted model')
```

```python
total_maxhr_set = np.arange(75,191)
```

```python
print(total_maxhr_set)
```

```python
total_maxhr_data = model.predict(total_maxhr_set.reshape(-1,1))
total_maxhr_data
```

```python
total_maxhr_factor = total_maxhr_data + 1
```

```python
print(total_maxhr_factor)
```
