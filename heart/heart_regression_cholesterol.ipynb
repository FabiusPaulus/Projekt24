```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  mean_squared_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('newdata\\new_heart_data.csv')
```

```python
x = data['Cholesterol'].values.reshape(-1, 1)
indices = np.where((x > 360) | (x == 0))

# Entferne Werte über 360
x = np.delete(x, indices).reshape(-1,1)

y = data['HeartDisease']
y = np.delete(y,indices)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y, color ='green')
plt.plot(x, model.predict(x), color='red')
plt.xlabel('cholesterol')
plt.ylabel('heart')
plt.title('fitted model')
```

```python
total_cholesterol_set = np.arange(100,351)
```

```python
print(total_cholesterol_set)
```

```python
total_cholesterol_data = model.predict(total_cholesterol_set.reshape(-1,1))
total_cholesterol_data
```
