```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  mean_squared_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('newdata\\new_heart_data.csv')
```

```python
data.columns
```

```python
x = data['Age'].values.reshape(-1, 1)
y = data['HeartDisease']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y, color ='green')
plt.plot(x, model.predict(x), color='red')
plt.xlabel('age')
plt.ylabel('heart')
plt.title('fitted model')
```

```python
total_age_set = np.arange(1,101)
```

```python
print(total_age_set)
```

```python
total_age_data = model.predict(total_age_set.reshape(-1,1))
total_age_data
```

```python
total_age_factor = total_age_data + 1
```

```python
print(total_age_factor)
```

```python
total_age_factor[80]
```

```python
age_age_factor_dict = {}
for i in range(len(total_age_set)):
    age_age_factor_dict[total_age_set[i]] = total_age_factor[i]
```
